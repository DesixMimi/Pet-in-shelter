```python
def transform_age_to_days(x):
            x = str(x)
            if x == 'nan':
                return np.nan
            elif 'year' in x:
                return int(x.split(' ')[0])*365
            elif 'month' in x:
                return int(x.split(' ')[0])*30
            elif 'week' in x:
                return int(x.split(' ')[0])*7
            elif 'day' in x:
                return int(x.split(' ')[0])*1
```

```python
def transform_age_to_years(x):
            x = str(x)
            if x == 'nan':
                return np.nan
            elif 'year' in x:
                return int(x.split(' ')[0])/1
            elif 'month' in x:
                return int(x.split(' ')[0])/12
            elif 'week' in x:
                return int(x.split(' ')[0])/52
            elif 'day' in x:
                return int(x.split(' ')[0])/365
```

```python
# DateTime - Remove the timestamp then split date into three new fields.
df.DateTime = pd.to_datetime(df.DateTime)
df["DayOfWeek"] = df.DateTime.dt.dayofweek
df["Month"] = df.DateTime.dt.month
df["year"] = df.DateTime.dt.year
```

```python
#mixed breed will be 1, else 0 in order to reduce the labels for future usage.
def breedType(x):
    if bool(re.search('/', x)):
        return 'Mix'
    elif bool(re.search('Mix', x)):
        return 'Mix'
    else:
        return 'Pure'
```

```python
def colorPurity(x):
    if bool(re.search('/', x)):
        return 'Mix'
    elif bool(re.search('Tricolor', x)):
        return 'Mix'
    else:
        return 'Pure'
```

```python
df['Age upon Outcome'] = df['Age upon Outcome'].apply(lambda x : transform_age_to_years(x))
df.rename(columns={'Age upon Outcome':'Age_upon_Outcome_years'},inplace=True)
df.Breed = df.Breed.apply(breedType)
df.Color = df.Color.apply(colorPurity)
```
